```python
from google.colab import files
import shutil


jobname = 'test' #@param {type:"string"}
sequence = '' #@param {type:"string"}
def make_fasta(jobname, query_sequence):
  with open(f'{jobname}.fasta', 'w') as f:
    f.write(f'>{jobname}\n{sequence}')
if sequence == '':
  print('Upload your own fasta file')
  uploaded = files.upload()
  for upload in uploaded.keys():
    shutil.copyfile(upload, f"{jobname}.fasta")
else:
  make_fasta(jobname, sequence)
```
